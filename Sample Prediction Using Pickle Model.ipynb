```python
import pickle
import pandas as pd
```

```python
with open("rainfall_model.pkl", "rb") as f:
    model = pickle.load(f)

# Create sample input
sample_input = pd.DataFrame([{
    "Location": "Sydney",
    "MinTemp": 13.4,
    "MaxTemp": 22.8,
    "Rainfall": 12.0,
    "Evaporation": 4.2,
    "Sunshine": 8.5,
    "WindGustDir": "NE",
    "WindGustSpeed": 35,
    "WindDir9am": "N",
    "WindDir3pm": "NE",
    "WindSpeed9am": 15,
    "WindSpeed3pm": 20,
    "Humidity9am": 95,
    "Humidity3pm": 80,
    "Pressure9am": 1015.2,
    "Pressure3pm": 1012.8,
    "Cloud9am": 8,
    "Cloud3pm": 8,
    "Temp9am": 16.5,
    "Temp3pm": 21.7,
    "RainToday": 1
}])

# IMPORTANT: Apply same encoding as training
sample_input = pd.get_dummies(sample_input)

# ⚠️ Now we need to match training columns
sample_input = sample_input.reindex(columns=model.feature_names_in_, fill_value=0)
```

```python

# Predict
prediction = model.predict(sample_input)

print("Prediction:", prediction)
```
